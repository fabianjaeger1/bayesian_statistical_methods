```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def estimate_pi(n):
    s = np.random.uniform(size=(n, 2))

    n_inside = np.sum(s[:,0]**2 + s[:,1]**2 < 1)

    return 4*n_inside/n

n = np.geomspace(10, 10000, 100, dtype=int)

plt.plot(n, [estimate_pi(n_) for n_ in n])
plt.axhline(np.pi)
plt.xscale("log")

plt.xlabel("$n$")
plt.ylabel(r"Estimate of $\pi$")
```
